# Chi2 Test

- カテゴリカル変数に対してカイ二乗検定した
- その結果をヒートマップにした

```python
import altair as alt
import pandas as pd
from pathlib import Path
```

```python
f_chi2 = "../../data/test_data/chi2_test/chi2_test.csv"
#f_p005 = "../../data/test_data/chi2_test/chi2_test_p005.csv"
```

```python
chi2_data = pd.read_csv(f_chi2)
#p005_data = pd.read_csv(f_p005)
```

```python
# def hmap(data: pd.DataFrame, scheme: str, p_limit: float):
#     mark = alt.Chart(data).mark_rect().encode(
#         alt.X("x"),
#         alt.Y("y"),
#         color = alt.condition(
#             alt.datum.p_value < p_limit,
#             alt.Color("p_value:Q").scale(domain=[0, p_limit], scheme=scheme, reverse=True),
#             alt.value("white")
#         )        
#     )
#     return mark

def hmap(data: pd.DataFrame, scheme: str, p_limit: float):
    q = f"p_value < {p_limit}"
    queried = data.query(q).copy()
    mark = alt.Chart(queried).mark_rect().encode(
        alt.X("x"),
        alt.Y("y"),
        alt.Color("p_value").scale(domain=[0, p_limit], scheme=scheme, reverse=True)
    )
    return mark
```

All

```python
p_limit = 1
scheme = "blues"
hm = hmap(chi2_data, scheme, p_limit)
hm
# fname = Path("../data/main_data/") / "chi2.png"
# hm.save(fname)
# print(f"Save chart to: {fname}")
```
