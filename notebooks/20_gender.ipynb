ジェンダーを軸にして、クロス集計する

```python
import pandas as pd
import altair as alt
import titanite as ti

print(f"Pandas {pd.__version__}")
print(f"Altair {alt.__version__}")
print(f"Titanite {ti.__version__}")
```

```python
fname = "../sandbox/config.toml"
cfg = ti.Config(fname=fname)
cfg.load()
category = cfg.categories()
cfg.questions
```

```python
fname = "../sandbox/tmp_preprocessed.csv"
data = pd.read_csv(fname, parse_dates=["timestamp"])
data = ti.categorical_data(data, category)
data.info()
```

```python
pd.crosstab(data["q1"], data["q2"])
```

```python
q = cfg.questions
def make_heatmap(data, col1, col2):
    key1 = col1.split("_")[0]
    key2 = col2.split("_")[0]

    cols = [col1, col2]
    grouped = data.groupby(cols)["response"].sum().reset_index()
    chart = alt.Chart(grouped).mark_rect().encode(
        alt.X(col1).title(q.get(key1)),
        alt.Y(col2),
        alt.Color("response").scale(scheme="blues"),
    ).properties(
        title=q.get(key2),
        width=500,
        height=500
    )
    return chart
```

```python
cols = []
for col in sorted(data.columns):
    if col in ("q2", "q15", "q16", "q18", "q20", "q21", "q22", "timestamp", "response"):
        pass
    else:
        cols.append(col)
cols
```

```python
for col in cols:
    make_heatmap(data, "q2", col).display()
```
